```python
import pdfplumber

cv_path = "CV Nassima_EL_HILALI_.pdf"

texte = ""
with pdfplumber.open(cv_path) as pdf:
    for page in pdf.pages:
        page_text = page.extract_text()
        if page_text:
            texte += page_text + "\n"

print(texte[:1500])
```

```python
sections = [
    "Compe´tences",
    "Expe´riences et Projets",
    "Formations"
]
```

```python
def decouper_blocs(texte, sections):
    blocs = {}
    lignes = texte.split("\n")

    current_section = None
    contenu = []

    for ligne in lignes:
        ligne_stripped = ligne.strip()

        # Si la ligne est un titre de section
        if ligne_stripped in sections:
            # Sauvegarder la section précédente
            if current_section is not None:
                blocs[current_section] = "\n".join(contenu).strip()

            # Nouvelle section
            current_section = ligne_stripped
            contenu = []
        else:
            if current_section:
                contenu.append(ligne)

    # Pour la dernière section
    if current_section:
        blocs[current_section] = "\n".join(contenu).strip()

    return blocs
```

```python
blocs = decouper_blocs(texte, sections)

# Afficher les titres détectés
list(blocs.keys())
```

```python
blocs_propres = {
    "Competences": blocs["Compe´tences"],
    "Experiences_et_Projets": blocs["Expe´riences et Projets"],
    "Formations": blocs["Formations"]
}

blocs_propres
```

```python
import pandas as pd

df_cv = pd.DataFrame([
    {"bloc": key, "texte": value}
    for key, value in blocs_propres.items()
])

df_cv
```

```python
import re
import unicodedata

def nettoyer_texte(t):
    # normaliser accents
    t = unicodedata.normalize("NFKD", t)
    # enlever les accents non reconnus
    t = t.encode("ascii", "ignore").decode("utf-8")
    # remplacer caractères collés
    t = re.sub(r"(?<=[a-z])(?=[A-Z])", " ", t)
    # enlever les caractères bizarres
    t = re.sub(r"[^a-zA-Z0-9.,;:?!\n\s]", " ", t)
    # remplacer multiples espaces
    t = re.sub(r"\s+", " ", t)
    return t.strip()
```
