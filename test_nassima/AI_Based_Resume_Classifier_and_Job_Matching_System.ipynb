```python
import PyPDF2

def extract_resume_text(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in range(len(reader.pages)):
            text += reader.pages[page].extract_text()
    return text

# Example usage:
resume_text = extract_resume_text('CV__Nassima_EL_HILALI.pdf')
print(resume_text)
```

```python
import spacy
import nltk
from nltk.corpus import stopwords
import string

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
nlp = spacy.load("en_core_web_sm")

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Tokenize and remove stopwords/punctuation
    doc = nlp(text)
    clean_tokens = [token.text for token in doc if token.text not in stop_words and token.text not in string.punctuation]

    return " ".join(clean_tokens)

# Example usage
cleaned_resume = preprocess_text(resume_text)
print(cleaned_resume)
```

```python
import streamlit as st
import PyPDF2
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import spacy
import string
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
nlp = spacy.load("en_core_web_sm")

# Function to extract resume text from a PDF
def extract_resume_text(file):
    reader = PyPDF2.PdfReader(file)
    text = ""
    for page in range(len(reader.pages)):
        text += reader.pages[page].extract_text()
    return text

# Preprocess text
def preprocess_text(text):
    text = text.lower()
    doc = nlp(text)
    clean_tokens = [token.text for token in doc if token.text not in stop_words and token.text not in string.punctuation]
    return " ".join(clean_tokens)

# Vectorize text and calculate cosine similarity
def calculate_similarity(resume, job_desc):
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([resume, job_desc])
    similarity_score = cosine_similarity(vectors)[0][1]
    return similarity_score

# Streamlit App
st.title("Resume Classifier and Job Matching System")

# Upload resume
uploaded_file = st.file_uploader("Upload your Resume (PDF)", type="pdf")

# Job description input
job_desc = st.text_area("Paste Job Description", height=150)

# If both resume and job description are provided
if uploaded_file and job_desc:
    resume_text = extract_resume_text(uploaded_file)
    cleaned_resume = preprocess_text(resume_text)
    cleaned_job_description = preprocess_text(job_desc)

    similarity_score = calculate_similarity(cleaned_resume, cleaned_job_description)

    # Display similarity score
    st.write(f"Resume-Job Description Similarity Score: {similarity_score:.4f}")
```

```python
import pandas as pd

df = pd.read_csv("UpdatedResumeDataSet.csv")
df.head()
```

```python
df.shape
```

Nettoyer les textes

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text

df["clean_resume"] = df["Resume"].apply(clean_text)
```

Offre

```python
job_description = """
Stage Data Scientist / Data Analyst
Analyse de donnees, machine learning, Python, statistiques, visualisation,
travail sur des donnees clients et dashboards.
"""

clean_job = clean_text(job_description)
```

TF-IDF entre tout les CV et l'offre et calcul de similarités cosine

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["clean_resume"].tolist() + [clean_job])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(X[:-1], X[-1])
df["similarity"] = similarities
```

Trier les CV du plus compatible au moins compatible

```python
df_sorted = df.sort_values(by="similarity", ascending=False)
df_sorted[["Category", "similarity"]].head(10)
```

Top 10 des CV les plus compatible

```python
import matplotlib.pyplot as plt

top10 = df_sorted[["Category", "similarity"]].head(10)

plt.figure()
plt.barh(top10["Category"], top10["similarity"])
plt.xlabel("Score de similarité")
plt.ylabel("Catégorie du CV")
plt.title("Top 10 des CV les plus compatibles avec l'offre")
plt.gca().invert_yaxis()   # le meilleur en haut
plt.show()
```
